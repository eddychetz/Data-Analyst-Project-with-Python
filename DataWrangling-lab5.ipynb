## Hands on Lab


Import pandas module.


```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

Load the dataset into a dataframe.


```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/Capstone_edX/Module%201/survey_results_public_2020.csv")
```

## Finding duplicates


In this section you will identify duplicate values in the dataset.


 Find how many duplicate rows exist in the dataframe.


```python
# your code goes here
df['is_duplicates'] = df.duplicated()
df.is_duplicates.value_counts()
```

## Removing duplicates


Remove the duplicate rows from the dataframe.


```python
# your code goes here
df = df.drop_duplicates()
```

Verify if duplicates were actually dropped.


```python
# your code goes here
df[df.duplicated()].head()
```

## Finding Missing values


Find the missing values for all columns.


```python
# your code goes here
df.apply(lambda x: sum(x.isnull().values), axis=0)
```

## Imputing missing values


Find the  value counts for the column Age.


```python
# your code goes here
df.Age.value_counts().to_frame()
```

Find the median for the column Age


```python
#your code goes here
median_age = df.Age.median()
median_age
```

Impute the median value to Age column


```python
# your code goes here
import numpy as np
df.Age.replace(np.nan, median_age,inplace=True)
```

Identify the value that is most frequent (majority) in the Country column.


```python
# your code goes here
df.Country.mode()
```

Drop all the missing values from the dataset


```python
# your code goes here
df_final = df.dropna()
df_final.info()
```

## Normalizing data


There are two columns in the dataset that talk about compensation.

One is "CompFreq". This column shows how often a developer is paid (Yearly, Monthly, Weekly).

The other is "CompTotal". This column talks about how much the developer is paid per Year, Month, or Week depending upon his/her "CompFreq". 

This makes it difficult to compare the total compensation of the developers.

In this section you will create a new column called 'NormalizedAnnualCompensation' which contains the 'Annual Compensation' irrespective of the 'CompFreq'.

Once this column is ready, it makes comparison of salaries easy.


<hr>


List out the various categories in the column 'CompFreq'


```python
# your code goes here
df_final.CompFreq.unique()
```

Create a new column named 'NormalizedAnnualCompensation'. Use the hint given below if needed.


Double click to see the **Hint**.

<!--

Use the below logic to arrive at the values for the column NormalizedAnnualCompensation.

If the CompFreq is Yearly then use the exising value in CompTotal
If the CompFreq is Monthly then multiply the value in CompTotal with 12 (months in an year)
If the CompFreq is Weekly then multiply the value in CompTotal with 52 (weeks in an year)

-->


```python
# your code goes here
df_final['NormalizedAnnualCompensation'] = np.where(df_final.CompFreq=='Monthly', df_final.CompTotal*12, df_final.CompTotal)
df_month=df_final.loc[df_final['CompFreq']=='Monthly']
df_month.head()
```

```python
df_final['NormalizedAnnualCompensation'] = np.where(df_final.CompFreq=='Weekly', df_final.CompTotal*52, df_final.NormalizedAnnualCompensation)
df_week = df_final.loc[df_final['CompFreq']=='Weekly']
df_week.head()
```

```python
df_final.merge(df_week, on='Respondent', how='inner')
```

```python
df_final.merge(df_month, on='Respondent', how='inner')
df_final[['CompFreq','CompTotal', 'NormalizedAnnualCompensation']].head()
```
