# Loading data

```
df = pd.read_csv('m5_survey_data_technologies_normalised.csv')
df = df[['Respondent', 'LanguageWorkedWith', 'LanguageDesireNextYear', 'DatabaseWorkedWith', 'DatabaseDesireNextYear', 'PlatformWorkedWith', 'PlatformDesireNextYear']]
df.head()
```

```
import matplotlib.pyplot as plt
df_lang = df['LanguageWorkedWith'].value_counts().to_frame()[0:5]
df_lang=df_lang.sort_values(by='LanguageWorkedWith')
df_lang.plot(kind='barh', figsize=(6,4), color='mbcYg', legend=False)
plt.title("Top 5 Current Language Trends")
plt.xlabel("Counts")
plt.ylabel("Language Worked With")
plt.xlim(0,10000)
plt.show()
```

```
import matplotlib.pyplot as plt
df_lang_nxt = df['LanguageDesireNextYear'].value_counts().to_frame()[0:5]
df_lang_nxt=df_lang_nxt.sort_values(by='LanguageDesireNextYear')
df_lang_nxt.plot(kind='barh', figsize=(6, 4), color='0cMyG', legend=False)
plt.title("Top 5 Language Desire Next Year")
plt.xlabel("Counts")
plt.ylabel("Language Desire Next Year")
plt.xlim(0,10000)
plt.show()
```
